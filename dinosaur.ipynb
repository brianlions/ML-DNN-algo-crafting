```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/datasaurus.csv', header=0)
names = df.dataset.unique()
plt.figure(figsize=(12,20), dpi=100)
for idx, name in enumerate(names):
    plt.subplot(5, 3, idx+1)

    part = df[ df.dataset == name ]
    x_mean, x_std = part.x.mean(), part.x.std()
    y_mean, y_std = part.y.mean(), part.y.std()
    rho = np.corrcoef(part.x, part.y)
    print('{:15s} {:.3f} {:.3f} {:.3f} {:.3f} {:.4f}'.format(name, x_mean, x_std, y_mean, y_std, rho[0,1]))
    plt.scatter(part.x, part.y, marker='.')
    #plt.xlim(0, 100)
    #plt.ylim(0, 100)
    plt.title(name)
```
